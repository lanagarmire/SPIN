```python
import anndata as ad
import numpy as np
import scanpy as sc
import sys
import os
import scipy
import pandas as pd

sys.path.append("D:\Research_project\For Test\smoothie\src")
from gaussian_smoothing import run_parallelized_smoothing
from spatial_correlation import compute_correlation_matrix
from network_analysis import make_spatial_network

input_path = r"D:\Research_project\For Test\smoothie\smoothie_input\celltype_specific\h5ad"

sample_id = [
    "H20.33.035", "H21.33.038", "H20.33.004", "H21.33.023", "H21.33.001", "H21.33.025", "H21.33.032",
    "H21.33.031", "H21.33.019", "H21.33.016", "H20.33.040", "H20.33.012", "H20.33.025", "H21.33.015",
    "H21.33.005", "H21.33.040", "H21.33.028", "H21.33.012", "H20.33.015", "H20.33.044", "H21.33.022",
    "H21.33.011", "H21.33.013", "H21.33.014", "H21.33.021", "H21.33.006"
]



for id in sample_id:   
        print(f"processing {id}...")
        # h5ad data are the same for different smoothie methods
        os.chdir(input_path)
        adata = ad.read_h5ad(f"{id}.h5ad")
        
        if id == "H21.33.032":
                min_spot = 5
                gaus_sd = 70

        else:
                min_spot = 20
                gaus_sd = 50
        


        sc.pp.normalize_total(adata, target_sum=1e3)
        sc.pp.log1p(adata)

        sm_adata = run_parallelized_smoothing(
            adata,
            grid_based_or_not=False,      
            gaussian_sd=gaus_sd,            
            min_spots_under_gaussian=min_spot
        )
        sm_adata = sm_adata[adata.obs_names].copy()

        smoothed_matrix = sm_adata.X.toarray() if scipy.sparse.issparse(sm_adata.X) else sm_adata.X
        df = pd.DataFrame(smoothed_matrix, index=sm_adata.obs_names, columns=sm_adata.var_names)

        os.chdir(r"D:\Research_project\For Test\smoothie\smoothie_output\celltype_specific\parameters_setting_1\csv\Astrocyte")
        df.to_csv(f"{id}_smoothed_expression_matrix.csv")


print("all patients complete")
```
